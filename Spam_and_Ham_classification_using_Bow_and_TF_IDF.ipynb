# Importing libraries and Dataset

```python
import pandas as pd
messages = pd.read_csv('SMSSpamCollection.txt', sep='\t',
                           names=["label", "message"])
```

```python
messages
```

# Data Cleaning and Preprocessing

```python
import nltk
import re
nltk.download('stopwords')
```

```python
# Importing libraries for stemming
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus= []
for i in range(0 , len(messages)):
  review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

# Creating  BOW (Bag of Words)






```python
# Create Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
# for binary Bow enable binary = true
cv = CountVectorizer(max_features=2500 , ngram_range=(1,2))
```

```python
x = cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems = 30, linewidth = 100000 , formatter= dict(float = lambda x: "%.3g" % x))
# Configures the way NumPy arrays are displayed in your notebook
"""
edgeitems = 30: This setting controls the number of elements displayed at the beginning and end of
each dimension when printing large arrays

linewidth = 100000: This determines the maximum number of characters allowed on a single line when
printing an array.

formatter=dict(float=lambda x: "%.3g" % x): This part customizes the formatting of floating-point
numbers within the array. The lambda function is used to format floats with three significant
digits using the "%.3g" format specifier.
"""
x
```

```python
# vocabulary
cv.vocabulary_
```

# Label Encoding

```python
# Output Features
y = pd.get_dummies(messages['label'])
y = y.iloc[:,0].values
```

```python
y
```

# Training and Testing the data

```python
# Train Test spilt
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 0)
```

```python
# Will be using Naive Bayes Classifier
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(x_train , y_train)
```

```python
spam_detect_model
```

```python
y_pred = spam_detect_model.predict(x_test)
```

# Checking the Accuracy and Classification Report

```python
# Importing the metrics
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
```

```python
accuracy_score(y_test , y_pred)
```

```python
classification_report(y_test , y_pred)
```

```python
confusion_matrix(y_test , y_pred)
```

# Creating TF IDF

```python
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20 , random_state=0)
```

```python
# Importing and creating TF IDF model
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X_train=tv.fit_transform(X_train).toarray()
X_test=tv.transform(X_test).toarray()
```

```python
tv.vocabulary_
```

# Using the Naive Bayes Classification

```python
spam_tfidf_model = MultinomialNB().fit(X_train, y_train)
```

```python
#prediction
y_pred=spam_tfidf_model.predict(X_test)
```

# Metrics for TF IDF

```python
accuracy_score(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
